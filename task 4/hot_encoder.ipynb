```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer

df = pd.read_excel("NarrateIQ_ML_Dataset.xlsx")
```

```python
data = df.copy()

categorical_cols = [
    "Age_Group", "Disability_Type", "Severity_Level", "State", "School_Type",
    "Language", "Teacher_Experience", "Input_Mode", "Content_Format", "Device_Type"
]

dummies = pd.get_dummies(data[categorical_cols], dtype=int)
merged = pd.concat([data.drop(columns=categorical_cols), dummies], axis="columns")

X = merged.drop(columns=["Student_ID", "Learning_Outcome", "Improvement_Score_Pct"])
y = merged["Improvement_Score_Pct"]

model = LinearRegression()
model.fit(X, y)

print(model.score(X, y))
```

```python
dfle = df.copy()

le = LabelEncoder()
for col in dfle.select_dtypes(include="object").columns:
    if col != "Learning_Outcome":
        dfle[col] = le.fit_transform(dfle[col])

dfle = df.copy()

le = LabelEncoder()
for col in dfle.select_dtypes(include="object").columns:
    if col != "Learning_Outcome":
        dfle[col] = le.fit_transform(dfle[col])

y = dfle["Improvement_Score_Pct"].values

ct = ColumnTransformer(
    [("town", OneHotEncoder(drop="first", sparse_output=False), [0])],
    remainder="passthrough"
)

X = ct.fit_transform(X)

model.fit(X, y)

X_pred = ct.transform(X[:1])
print(model.predict(X_pred))  
```
