```python
import pandas as pd
import numpy as np
from sklearn import linear_model

file_path = "NarrateIQ_ML_Dataset.xlsx"
df = pd.read_excel(file_path)
df.head()
```

```python
X = df[["Engagement_Score"]]
y = df["Improvement_Score_Pct"]

# Train the linear regression model
model = linear_model.LinearRegression()
model.fit(X, y)

# Check model parameters and make a test prediction
print("Coefficient:", model.coef_)
print("Intercept:", model.intercept_)

sample = pd.DataFrame({"Engagement_Score": [70]})
print("Prediction for Engagement_Score = 70:", model.predict(sample))
```

```python
import pickle
with open('model_pickle.pkl', 'wb') as file:
    pickle.dump(model, file)

with open('model_pickle.pkl', 'rb') as file:
    mp = pickle.load(file)
print('Loaded Coefficient:', mp.coef_)
print('Loaded Intercept:', mp.intercept_)
print('Loaded Prediction for Engagement_Score = 70:', mp.predict(pd.DataFrame({'Engagement_Score': [70]})))
```

```python
import joblib
joblib.dump(model, 'model_joblib.joblib')
mj = joblib.load('model_joblib.joblib')
print('Loaded Coefficient:', mj.coef_)
print('Loaded Intercept:', mj.intercept_)
print('Loaded Prediction for Engagement_Score = 70:', mj.predict(pd.DataFrame({'Engagement_Score': [70]})))

```
